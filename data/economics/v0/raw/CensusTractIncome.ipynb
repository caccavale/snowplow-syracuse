```python
import pandas as pd
import os
```

```python
econ_stats = pd.read_csv('ACS_16_5YR_DP03.csv')
```

```python
econ_stats.head()
```

```python
new_names = list(filter(lambda x: 'HC03' in x or 'GEO' in x, econ_stats))
```

```python
new_names = new_names[0:3] + new_names[new_names.index('HC03_VC75'):new_names.index('HC03_VC84') + 1]
new_names
```

```python
just_income = econ_stats[new_names]
just_income.head()
```

```python
just_income.columns = just_income.iloc[0]
just_income = just_income[1:]
just_income.head()
```

```python
just_income.columns[10][87:94]
```

```python
just_income.rename(columns= lambda x: x if x in ['Id', 'Id2', 'Geography'] else x[87:], inplace=True)
```

```python
just_income["Id2"] = just_income["Id2"].apply(lambda x: int(x))
```

```python
cuse_tracts = pd.read_csv('Census_Tracts_in_Syracuse_NY_2010.csv')
```

```python
syr_income = just_income[[x in list(cuse_tracts.GEOID10) for x in just_income.Id2]]
```

```python
syr_income.to_csv('../clean/census_tract_income.csv')
```
